# Úvod

## Co je TAS?

-   tripple axis spectrometer
-   funguje na neutronových zdrojích
-   velmi drahý měřící čas
-   v nepružném módu detekuje jednotky neutronů za minutu, měření jsou
    tedy velmi dlouhá
-   [ukázka
    zde](https://www.ill.eu/users/support-labs-infrastructure/software-scientific-tools/neutrons4science)

## Mapování reciprokého prostoru

-   měření jednoho bodu trvá nenulovou dobu
-   přesun mezi body také trvá (cost function)
-   data obsahují šum



## Nestandardní jádra

-   RBF neumí popsat data, kde se něco děje jen někde (je statický pro
    celý obrázek)
-   řešení:
    -   parametrická jádra
    -   deep kernel



# gpCAM

## gpCAM - coto je?

-   Autonomous Data Acquisition, Uncertainty Quantification and HPC
    Optimization
-   vysoká flexibilita, je možné zadat:
    -   data acquisition function
    -   kernel function
    -   parametrická prior funkce
    -   vlastní optimizer
    -   cost function
